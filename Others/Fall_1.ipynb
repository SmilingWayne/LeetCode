```python
import heapq

def find_top_k(arrays, k):
    """
    在 n 个升序数组中找到最小的 k 个元素
    
    参数:
    arrays: 包含 n 个升序数组的列表
    k: 要查找的元素数量
    
    返回:
    包含最小 k 个元素的列表
    """
    n = len(arrays)
    if n == 0 or k <= 0:
        return []
    
    # 初始化最小堆
    min_heap = []
    
    # 将每个数组的第一个元素及其位置信息加入堆中
    for i in range(n):
        if arrays[i]:  # 确保数组非空
            # (元素值, 数组索引, 元素在数组中的索引)
            heapq.heappush(min_heap, (-arrays[i][-1], i, len(arrays[i]) - 1))
    
    result = []
    
    # 从堆中弹出最小元素，直到找到 k 个元素或堆为空
    while min_heap and len(result) < k:
        val, arr_idx, elem_idx = heapq.heappop(min_heap)
        result.append(val)
        
        # 如果当前数组还有更多元素，将下一个元素加入堆中
        if elem_idx - 1 >= 0:
            next_val = arrays[arr_idx][elem_idx - 1]
            heapq.heappush(min_heap, (- next_val, arr_idx, elem_idx - 1))
    
    return list(map(lambda x: -x, result))

# 测试示例
if __name__ == "__main__":
    # 示例1
    arrays1 = [
        [1, 4, 7, 10],
        [2, 5, 8, 11],
        [3, 6, 9, 12]
    ]
    k1 = 5
    print(f"Top {k1} in arrays1: {find_top_k(arrays1, k1)}")
    # 输出: [1, 2, 3, 4, 5]
    
    # 示例2
    arrays2 = [
        [10, 20, 30, 40],
        [15, 25, 35],
        [27, 29, 37, 48, 50],
        [32, 33]
    ]
    k2 = 7
    print(f"Top {k2} in arrays2: {find_top_k(arrays2, k2)}")
    # 输出: [10, 15, 20, 25, 27, 29, 30]
    
    # 示例3 (空数组处理)
    arrays3 = [
        [1, 3, 5],
        [],
        [2, 4, 6]
    ]
    k3 = 4
    print(f"Top {k3} in arrays3: {find_top_k(arrays3, k3)}")
    # 输出: [1, 2, 3, 4]
```
